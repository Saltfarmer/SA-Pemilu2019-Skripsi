```python
%%time
import pandas as pd
import numpy as np
line = pd.read_csv('Line-Today-Label/line_label.csv')
# pbw = pd.read_csv('Line-Today-Label\prabowo_subianto_label1.csv')
# line = pd.concat([jkw, pbw])
```

```python
%%time
line.dropna(subset=['Berita'], inplace=True)
```

```python
%%time
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import cross_val_score

x = line.Berita
y = line.Sentimen

xcv = x
xtf = x

countvect = CountVectorizer(analyzer = "word", tokenizer = None, lowercase = None)
tfidf = TfidfVectorizer(analyzer = "word", tokenizer = None, lowercase = None)

xcv = countvect.fit_transform(x).toarray()
xtf = tfidf.fit_transform(x).toarray()
```

```python
%%time
from sklearn.feature_selection import chi2, f_classif, mutual_info_classif, SelectPercentile

xcvchi = SelectPercentile(chi2, percentile = 50).fit_transform(xcv, y)
xcvano = SelectPercentile(f_classif, percentile = 50).fit_transform(xcv, y)
xcvmi = SelectPercentile(mutual_info_classif, percentile= 50).fit_transform(xcv, y)

xtfchi = SelectPercentile(chi2, percentile = 50).fit_transform(xtf, y)
xtfano = SelectPercentile(f_classif, percentile = 50).fit_transform(xtf, y)
xtfmi = SelectPercentile(mutual_info_classif, percentile= 50).fit_transform(xtf, y)
```

```python
%%time
seed = 17
import numpy as np
'''#Random Forest Classifier'''
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100, random_state=seed)

'''Support Vector Machines'''
from sklearn.svm import SVC
svm = SVC(gamma = 'auto', kernel='linear',  random_state=seed)

'''Decision Tree Classifier'''
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=seed)

'''Extra Tree Classifier'''
from sklearn.ensemble import ExtraTreesClassifier
etc = ExtraTreesClassifier(n_estimators=100, random_state=seed)

'''K-Nearest Neighbour'''
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()

'''Multinomial Naive Bayes'''
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()

'''Gaussian Naive Bayes'''
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()

'''Logistic Regression'''
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(solver='saga', random_state=seed)

'''Neural Network'''
from sklearn.neural_network import MLPClassifier
nn = MLPClassifier(random_state=seed)

'''ADA Boosting'''
from sklearn.ensemble import AdaBoostClassifier
abc = AdaBoostClassifier(n_estimators=100, random_state=seed)

'''Gradient Boosting'''
from sklearn.ensemble import GradientBoostingClassifier
gbc = GradientBoostingClassifier(n_estimators=100, random_state=seed)
```

```python
%%time
import time

def x_val_accuracy(model,x,y):
    from sklearn.model_selection import cross_val_score
    x_val_score = cross_val_score(model, x, y, cv = 10, scoring = 'accuracy').mean()
    x_val_score = np.round(x_val_score*100, 2)
    return x_val_score

def x_val_precision(model,x,y):
    from sklearn.model_selection import cross_val_score
    x_val_score = cross_val_score(model, x, y, cv = 10, scoring = 'precision').mean()
    x_val_score = np.round(x_val_score*100, 2)
    return x_val_score

def x_val_recall(model,x,y):
    from sklearn.model_selection import cross_val_score
    x_val_score = cross_val_score(model, x, y, cv = 10, scoring = 'recall').mean()
    x_val_score = np.round(x_val_score*100, 2)
    return x_val_score

def x_val_f1(model,x,y):
    from sklearn.model_selection import cross_val_score
    x_val_score = cross_val_score(model, x, y, cv = 10, scoring = 'f1').mean()
    x_val_score = np.round(x_val_score*100, 2)
    return x_val_score

def runtime(model,x,y):
    start = time.time()
    from sklearn.model_selection import cross_val_score
    x_val_score = cross_val_score(model, x, y, cv = 10, scoring = 'accuracy').mean()
    x_val_score = np.round(x_val_score*100, 2)
    runtime = time.time() - start
    return np.round(runtime, 2)
```

```python
%%time
x_val_cv = pd.DataFrame({'accuracy(%)':[x_val_accuracy(rf,xcv,y), x_val_accuracy(svm,xcv,y), x_val_accuracy(dt,xcv,y), x_val_accuracy(etc,xcv,y), x_val_accuracy(knn,xcv,y), x_val_accuracy(mnb,xcv,y), x_val_accuracy(gnb,xcv,y), x_val_accuracy(lr,xcv,y), x_val_accuracy(nn,xcv,y), x_val_accuracy(abc,xcv,y), x_val_accuracy(gbc,xcv,y)],\
                        'precision(%)':[x_val_precision(rf,xcv,y), x_val_precision(svm,xcv,y), x_val_precision(dt,xcv,y), x_val_precision(etc,xcv,y), x_val_precision(knn,xcv,y), x_val_precision(mnb,xcv,y), x_val_precision(gnb,xcv,y), x_val_precision(lr,xcv,y), x_val_precision(nn,xcv,y), x_val_precision(abc,xcv,y), x_val_precision(gbc,xcv,y)],\
                        'recall(%)':[x_val_recall(rf,xcv,y), x_val_recall(svm,xcv,y), x_val_recall(dt,xcv,y), x_val_recall(etc,xcv,y), x_val_recall(knn,xcv,y), x_val_recall(mnb,xcv,y), x_val_recall(gnb,xcv,y), x_val_recall(lr,xcv,y), x_val_recall(nn,xcv,y), x_val_recall(abc,xcv,y), x_val_recall(gbc,xcv,y)],\
                        'f1(%)':[x_val_f1(rf,xcv,y), x_val_f1(svm,xcv,y), x_val_f1(dt,xcv,y), x_val_f1(etc,xcv,y), x_val_f1(knn,xcv,y), x_val_f1(mnb,xcv,y), x_val_f1(gnb,xcv,y), x_val_f1(lr,xcv,y), x_val_f1(nn,xcv,y), x_val_f1(abc,xcv,y), x_val_f1(gbc,xcv,y)], \
                        'time(s)':[runtime(rf,xcv,y), runtime(svm,xcv,y), runtime(dt,xcv,y), runtime(etc,xcv,y), runtime(knn,xcv,y), runtime(mnb,xcv,y), runtime(gnb,xcv,y), runtime(lr,xcv,y), runtime(nn,xcv,y), runtime(abc,xcv,y), runtime(gbc,xcv,y)]})
x_val_cv.index = ['RF', 'SVM', 'DT', 'ETC', 'KNN', 'MNB', 'GNB', 'LR', 'NN', 'ABC', 'GBC']

print(x_val_cv)
x_val_cv.to_csv(r'x_val_cv.csv', header=True)
```
